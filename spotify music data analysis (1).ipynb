```python
sp['duration'] = sp['duration_ms'].apply (lambda x : round(x/1000))
sp.drop('duration_ms', inplace = True, axis=1)
sp.duration.head()
```

```python
numeric_data = sp.select_dtypes(include=['number'])
td = numeric_data.drop(['key', 'mode', 'explicit'], axis=1, errors='ignore').corr()
plt.figure(figsize=(10, 6))
hmap = sns.heatmap(td, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0,
                   cmap='Greens', linewidths=0.1, linecolor='black')
hmap.set_title('Correlation HeatMap', fontsize=14)
hmap.set_xticklabels(hmap.get_xticklabels(), rotation=90)
plt.tight_layout()
plt.show()
```
